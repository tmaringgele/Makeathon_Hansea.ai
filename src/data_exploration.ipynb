```python
# Read the CSV file into a dataframe
df_class_dict = pd.read_csv('../data/segmentation/archive/class_dict.csv')

# Display the dataframe
df_class_dict
```

```python
# Read the CSV file into a dataframe
df_metadata = pd.read_csv('../data/segmentation/archive/metadata.csv')

# Display the dataframe
df_metadata
```

```python
# Split the dataframe into training and test/val dataframes
df_train = df_metadata.loc[df_metadata['split'] == 'train']
df_test = df_metadata.loc[df_metadata['split'] == 'test']

df_train = df_train.drop(['image_id', 'split'], axis=1)
df_test = df_test.drop(['image_id', 'split'], axis=1)


# Save the training and test/val dataframes as CSV files
df_train.to_csv('../data/segmentation/archive/train/train_data.csv', index=False)
df_test.to_csv('../data/segmentation/archive/test/test_data.csv', index=False)
```
