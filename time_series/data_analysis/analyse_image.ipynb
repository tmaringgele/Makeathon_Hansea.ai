```python
from PIL import Image
import numpy as np
import os
import json

```

```python
def getm2(image_tiff):
    imarray = np.asarray(Image.open(image_tiff))
    return sum(sum(imarray)) * 400


def analyzeDir(directory):
    quadrant = []
    for filename in os.listdir(directory):
        f = os.path.join(directory, filename).replace("\\", "/")
        # checking if it is a file
        if os.path.isfile(f):
            if "CONF" not in f and "MASK" not in f:
                quadrant.append((f.split("_")[1], getm2(f)))
    return quadrant


directory = 'data'

result = []
# Liste der Unterverzeichnisse im angegebenen Verzeichnis
subdirectories = [d for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]

# Schleife über alle Unterverzeichnisse
for subdir in subdirectories:
    # Vollständiger Pfad zum aktuellen Unterverzeichnis
    subdir_path = os.path.join(directory, subdir).replace("\\", "/")
    # Hier können Sie mit dem Unterverzeichnis arbeiten, z. B. seine Inhalte durchlaufen
    result.append((subdir_path.split("/")[1],analyzeDir(subdir_path)))

        
print(result)

# File path to save the JSON file
file_path = 'output.json'

# Save the Python array as JSON
with open(file_path, 'w') as json_file:
    json.dump(result, json_file)

```
